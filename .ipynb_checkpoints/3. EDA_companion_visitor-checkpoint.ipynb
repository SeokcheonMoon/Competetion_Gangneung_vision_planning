## 데이터 탐색

### 1. 동반 유형

```python
df_companion["동반유형명"].unique()
```

```python
df_companion = df_companion[df_companion['동반유형명'] != '기타가족']
df_companion
```

```python
df_companion['연도'] = df_companion['기준연월'].astype(str).str[:4]
df_companion['월'] = df_companion['기준연월'].astype(str).str[4:6]
df_companion = df_companion.sort_values(by = "기준연월")
df_companion
```

```python
df_2022_companion = df_companion.query("연도 == '2022'")
df_2023_companion = df_companion.query("연도 == '2023'")
```

```python
df_2022_companion
```

```python
# 월별로 언급량수의 최대값을 가지는 목적지 유형 찾기
max_mentions = df_2022_companion.loc[df_2022_companion.groupby(['월'])['언급건수'].idxmax()]

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(max_mentions['월'], max_mentions['언급건수'], color='magenta')
plt.xlabel('월')
plt.ylabel('최대 언급건수')
plt.title('2022년 월별 최대 언급건수에 따른 동반유형명')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(max_mentions)):
    plt.text(i, max_mentions['언급건수'].iloc[i], max_mentions['동반유형명'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# 월별로 언급량수의 최대값을 가지는 목적지 유형 찾기
max_mentions = df_2023_companion.loc[df_2023_companion.groupby(['월'])['언급건수'].idxmax()]

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(max_mentions['월'], max_mentions['언급건수'], color='magenta')
plt.xlabel('월')
plt.ylabel('최대 언급건수')
plt.title('2023년 월별 최대 언급건수에 따른 동반유형명')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(max_mentions)):
    plt.text(i, max_mentions['언급건수'].iloc[i], max_mentions['동반유형명'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
