### 2. 목적지 검색량

```python
df_destination_search
```

```python
df_destination_search = df_destination_search[df_destination_search['목적지 유형'] != '전체']
df_destination_search
```

```python
df_destination_search['연도'] = df_destination_search['기준연월'].astype(str).str[:4]
df_destination_search['월'] = df_destination_search['기준연월'].astype(str).str[4:6]
df_destination_search = df_destination_search.sort_values(by = "기준연월")
df_destination_search
```

```python
df_2022_destination_search = df_destination_search.query("연도 =='2022'")
df_2023_destination_search = df_destination_search.query("연도 =='2023'")
```

```python
# 월별로 언급량수의 최대값을 가지는 목적지 유형 찾기
max_mentions = df_2022_destination_search.loc[df_2022_destination_search.groupby(['월'])['목적지 검색량'].idxmax()]

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(max_mentions['월'], max_mentions['목적지 검색량'], color='skyblue')
plt.xlabel('월')
plt.ylabel('최대 목적지 검색량')
plt.title('2022년 월별 최대 목적지 검색량에 따른 목적지 유형')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(max_mentions)):
    plt.text(i, max_mentions['목적지 검색량'].iloc[i], max_mentions['목적지 유형'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
second_max_mentions = df_2022_destination_search.groupby('월').apply(lambda x: x.nlargest(2, '목적지 검색량').iloc[-1]).reset_index(drop=True)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(second_max_mentions['월'], second_max_mentions['목적지 검색량'], color='skyblue')
plt.xlabel('월')
plt.ylabel('최대 목적지 검색량')
plt.title('2022년 월별 두번째로 높은 목적지 검색량에 따른 목적지 유형')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(second_max_mentions)):
    plt.text(i, second_max_mentions['목적지 검색량'].iloc[i], second_max_mentions['목적지 유형'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# 월별로 언급량수의 최대값을 가지는 목적지 유형 찾기
max_mentions = df_2023_destination_search.loc[df_2023_destination_search.groupby(['월'])['목적지 검색량'].idxmax()]

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(max_mentions['월'], max_mentions['목적지 검색량'], color='skyblue')
plt.xlabel('월')
plt.ylabel('최대 목적지 검색량')
plt.title('2023년 월별 최대 목적지 검색량에 따른 목적지 유형')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(max_mentions)):
    plt.text(i, max_mentions['목적지 검색량'].iloc[i], max_mentions['목적지 유형'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
second_max_mentions = df_2023_destination_search.groupby('월').apply(lambda x: x.nlargest(2, '목적지 검색량').iloc[-1]).reset_index(drop=True)

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(second_max_mentions['월'], second_max_mentions['목적지 검색량'], color='skyblue')
plt.xlabel('월')
plt.ylabel('최대 목적지 검색량')
plt.title('2023년 월별 두번째로 높은 목적지 검색량에 따른 목적지 유형')
plt.xticks(rotation=45)
plt.grid(axis='y')

# 목적지 유형 텍스트 추가
for i in range(len(second_max_mentions)):
    plt.text(i, second_max_mentions['목적지 검색량'].iloc[i], second_max_mentions['목적지 유형'].iloc[i], ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## 분석 결과 및 인사이트 도출

#### 1. 관광 명소 정보
- 자연경관 감상 관련 명소 개수가 1위를 차지.
- 호텔,콘도,모텔 등을 숙박업소로 하나로 묶을 때 숙박업소의 개수가 2위를 차지.

#### 2. 목적지 검색량
- 2022,2023년 모두 사람들이 음식과 관련된 목적지 검색을 많이함.
- 계절과 관계없이 대체적으로 음식, 숙박, 자연관광이 높은 검색량을 차지함.

#### 3. 인기 맛집
- 인기 맛집 조사 결과 한식, 카페, 간이음식, 전문음식, 외국식 등 음식점의 개수가 확연한 차이를 보임.

#### Insight
- 숙박에 관심도가 높은 만큼 숙박에 대한 관광 명소가 많이 마련되어있음.
- 음식과 관련된 데이터 분석 후 음식점 추천 아이템 구상하기.
